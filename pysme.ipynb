```python
import pandas as pd
from io import StringIO

input_file = "KennyPhan.013192"
output_file = "sorted_linelist.lin"

# --- Step 1: Read all lines ---
with open(input_file, "r") as f:
    lines = f.readlines()

# --- Step 2: Split into sections ---
header = lines[:4]               # first 4 lines stay unchanged
data_lines = lines[4:30134]      # lines to sort
footer = lines[30134:]           # keep the remainder untouched

# --- Step 3: Load data section into a DataFrame ---
df = pd.read_csv(StringIO("".join(data_lines)), sep=",", engine="python")

# --- Step 4: Sort by wavelength column ---
df_sorted = df.sort_values(by="WL_air(A)", ascending=True)

# --- Step 5: Write to new file ---
with open(output_file, "w") as f:
    f.writelines(header)               # write header
    df_sorted.to_csv(f, index=False)   # write sorted section
    f.writelines(footer)               # append the rest unchanged

print(f"Sorted file saved as {output_file}")
```
